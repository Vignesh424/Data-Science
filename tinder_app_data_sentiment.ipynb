```python
df = pd.read_csv("link to file", encoding='utf8')

df
```

```python
# Adds sentiment columns | Credits to Stack Overflow
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None

df['sentiment'] = df['content'].apply(sentiment_calc)
```

```python
# Adds polarity column | Credits to Stack Overflow
def polarity_calc(text):
    try:
        return TextBlob(text).polarity
    except:
        return None

df['polarity'] = df['content'].apply(polarity_calc)
```

```python
# Adds subjectivity column | Credits to Stack Overflow
def subjectivity_calc(text):
    try:
        return TextBlob(text).subjectivity
    except:
        return None

df['subjectivity'] = df['content'].apply(subjectivity_calc)
```
